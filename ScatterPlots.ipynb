```python
plt.figure(figsize=(4,4), dpi = 200)
plt.title("Scatter Plot of Distribution on Exam Score and Time Spent to Study", fontsize = 12, color = "blue", fontweight = "bold")
sns.scatterplot(data = df, x = "Hours_Studied", y = "Exam_Score", legend=True, color = "blue");

plt.xlabel("Hours_Studied", fontweight = "bold")
plt.ylabel("Exam_Score", fontweight = "bold")
```
